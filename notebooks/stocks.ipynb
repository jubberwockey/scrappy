```python
import pandas_datareader as pdr
# import yahoo_finance as yf
# import yahoofinancials
# import quandl as ql

import seaborn as sns
```

```python
start = '1990-01-01'
end = '2020-07-18'
ticker = 'nasdaqcom'

x = pdr.DataReader([ticker], 'fred', start, end)
x['diff'] = x[[ticker]] - x[[ticker]].shift(-1)
x['returns'] = x[[ticker]]/x[[ticker]].shift(-1) - 1.
```

```python
axis = 'diff'

sns.distplot(x[axis])

print(x[axis].mean(), x[axis].median(), x[axis].std(), x[axis].skew(), x[axis].kurt())
x[axis].mode()
```
